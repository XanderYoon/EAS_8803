# 1. Environment Setup

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Clustering algorithms
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn_extra.cluster import KMedoids
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('hurricane_data.csv')
df.head()
```

## 1.1 Plot Original Storms

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    x= df['Rainfall_Max'],
    y= df['Damages'],
    c=df['Storm'],
    cmap='Accent'
)

# Add color bar and labels
plt.xlabel('Maximum Rainfall')
plt.ylabel('Damages')
plt.title('Scatter Plot by Storm')
plt.show()
```

```python
def plotting_function(Algorithim, labels):
    # Plot K-Means Clustering Results
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(
        x= df['Rainfall_Max'],
        y= df['Damages'],
        c= labels,
        cmap='viridis'
    )

    # Add color bar and labels
    plt.xlabel('Maximum Rainfall')
    plt.ylabel('Damages')
    plt.title(Algorithim)
    plt.show()

X = df[['Rainfall_Max', 'Damages']]
```

# 2. K-Means

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
plotting_function("K-Means", kmeans_labels)
```

# 3. K-Medioids

```python
# K-Medoids Clustering
# You may need to install scikit-learn-extra if not already installed:
# !pip install scikit-learn-extra

kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids_labels = kmedoids.fit_predict(X)
plotting_function("K-Medioids", kmedoids_labels)
```

# 4. DBSCAN

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
plotting_function("DBSCAN", dbscan_labels)
```

# 5. Agglomerative

```python
# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X)
plotting_function("Agglomerative", agglo_labels)
```

# 6. Gaussian Mixture Model (GMM)

```python
# Gaussian Mixture Model Clustering
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X)
plotting_function("GMM", gmm_labels)
```
